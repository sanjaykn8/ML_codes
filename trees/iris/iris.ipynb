```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split

df = load_iris()
Xc, yc = df.data, df.target

Xc_train, Xc_test, yc_train, yc_test = train_test_split(Xc, yc, random_state=0, test_size=0.2, stratify=yc)
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=0)
clf.fit(Xc_train, yc_train)
```

```python
plt.figure(figsize=(8,6))
plot_tree(clf, feature_names=df.feature_names, class_names=df.target_names, filled=True)
plt.title("Decision Tree (Iris)")
plt.show()
```

```python
accuracy = clf.score(Xc_test, yc_test)
print(f"Accuracy: {accuracy:.4f}")
```

```python
target = [[5.1,3.5,1.4,0.2]]
pred = clf.predict(target)[0]
print(f"Predicted class for {target[0]}: {df.target_names[pred]}")
```
